# **Dummies Method**

get_dummies

```python
cds=ds.copy()
```

```python
dummies=pd.get_dummies(cds["Gender"],prefix='Gender_')
```

```python
dummies
```

```python
dummies=dummies.astype(int)
```

```python
dummies
```

```python
dp=pd.concat([cds,dummies],axis=1)     #axis=1 mean column wise
```

# **One Hot Encoder**

```python
ds['Married'].fillna(ds['Married'].mode()[0],inplace=True)
ds.isnull().sum()
```

```python
new_data=ds[['Gender','Married']]
```

```python
new_data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder() #creating object of one hot encoder
```

```python
ar=ohe.fit_transform(new_data).toarray()
```

```python
ar
```

```python
pd.get_dummies(new_data).info()
```

```python
after_label_ar=pd.DataFrame(ar,columns=["Gender_Female","Gender_Male","Married_NO","Married_Yes"])
after_label_ar
```
